```python
## ElGamal Enc ##
#pick a at random and compute A = g^a
print("------Encryption Step------")
a = randint(1,q)
A = g^a
print('a =', a)
print('A =', A)
#compute secret key K
K1 = B^a
print('K1 =', K1)
#encrypt message M with K
M = Zp(114514)
C = K1*M
print('M = ', M)
print('C = Enc(M) = ', C)
```

```python
## ElGamal Dec ##
#compute secret key K
print("------Decryption Step------")
K2 = A^b
print('K2 =', K2)
#encrypt message M with K
M = C/K2
print('M = Dec(C) = ', M)
```
