```python
########## Textbook RSA Signature ##################
### Textbook RSA Signature KeyGen ####
n_bits = 2048
p = random_prime(2^(n_bits/2))
q = random_prime(2^(n_bits/2))
N = p*q
e = 2^16+1
print('Verification key is')
print('N =',N)
print('e =',e)

print('Signing key is')
d = lift(mod(e,(p-1)*(q-1))^(-1))
print('N =',N)
print('d =',d)
Zp=Integers(N)# calculate over Zp
```

```python
####### Textbook RSA Signature Sign ###########
M = mod(114514, N)

S = M^d
print('M =', M)
print('S = Sign(M)', S)
```

```python
####### Textbook RSA Signature Vrfy ###########
Mprime = S^e
print('M\' =',Mprime)
print('Did verification succeed?')
print('Vrfy(S) =',Mprime==M)
```

```python
####### Attack of Textbook RSA Signature ###########
#Sign tow messages M1 M2
M1 = mod(114, N)
M2 = mod(514, N)

S1 = M1^d
S2 = M2^d

print('M1, M2 =', M1, ',', M2)
print('S1, S2 = Sign(M1), Sign(M2) = ', S1, S2)

#Forge a new message based on M1 M2
M12 = M1*M2
S12 = S1*S2
print('M1*M2 =', M12)
print('S1*S2 = Sign(M1)*Sign(M2) = ', S12)

#Verify the forged message
M12prime = S12^e
print('M12\' =',M12prime)
print('Did verification succeed? (i.e., attack succeeded)')
print('Vrfy(S12) =',M12prime==M12)
```
